```python
data = pd.read_csv("weather.csv")
data
```

```python
X = data.drop("RainTomorrow", axis=1)
y = data['RainTomorrow']
```

```python
data = data.fillna(0)
```

```python
data.info()
```

```python
from sklearn import preprocessing
```

```python
convert = preprocessing.LabelEncoder()
X['RainToday'] = convert.fit_transform(X['RainToday'])
y = convert.fit_transform(y)
X
```

```python
convert = preprocessing.LabelEncoder()
X['WindGustDir'] = convert.fit_transform(X['WindGustDir'])
y = convert.fit_transform(y)
X
```

```python
convert = preprocessing.LabelEncoder()
X['WindDir9am'] = convert.fit_transform(X['WindDir9am'])
X['WindDir3pm'] = convert.fit_transform(X['WindDir3pm'])
y = convert.fit_transform(y)
X
```

```python
from sklearn import preprocessing, metrics
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB 
%matplotlib inline 
```

```python
gNB = GaussianNB()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30) 

print("There are {} samples in the training dataset".format(X_train.shape[0]))
print("There are {} samples in the testing dataset".format(X_test.shape[0]))
print("Each sample has {} features".format(X_train.shape[1]))
```

```python
from sklearn.impute import SimpleImputer

# Assuming X_train and X_test have NaN values
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# Now train your GaussianNB model on X_train_imputed
NB = GaussianNB().fit(X_train_imputed, y_train)
predicted_y = NB.predict(X_test_imputed)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics

# Assuming you have X_train, X_test, y_train, and y_test ready
# Assuming X_train contains NaN values

# Create an imputer to fill NaN values with the mean (you can choose a different strategy)
imputer = SimpleImputer(strategy='mean')

# Fit the imputer on the training data and transform X_train
X_train_imputed = imputer.fit_transform(X_train)

# Create and train the Gaussian Naive Bayes model
modifyNB = GaussianNB().fit(X_train_imputed, y_train)

# Transform X_test using the same imputer
X_test_imputed = imputer.transform(X_test)

# Make predictions on the test set
predicted_y = modifyNB.predict(X_test_imputed)

# Evaluate the model
accuracy = metrics.accuracy_score(y_test, predicted_y)
average_accuracy = np.mean(y_test == predicted_y) * 100
print("The average_accuracy is {0:.1f}%".format(average_accuracy))

# Display classification report
print(metrics.classification_report(y_test, predicted_y))

# Display confusion matrix
confusion_matrix = metrics.confusion_matrix(y_test, predicted_y)
print("Confusion Matrix:")
print(confusion_matrix)
```
